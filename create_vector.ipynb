```python
# Import pustaka
import numpy as np
import matplotlib.pyplot as plt

# Buat grid untuk x dan y
x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))

# === Soal (a): F(x, y) = x*i + (1/2)y*j ===
Fx_a = x
Fy_a = 0.5 * y

# Plot medan vektor soal (a)
plt.figure(figsize=(8, 8))
plt.quiver(x, y, Fx_a, Fy_a, angles='xy', scale_units='xy', scale=1, color='blue')
plt.title('Medan Vektor (a): $\\mathbf{F}(x, y) = x\\mathbf{i} + \\frac{1}{2}y\\mathbf{j}$', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$y$', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()

# === Soal (b): F(x, y) = y*i + (x + y)*j ===
Fx_b = y
Fy_b = x + y

# Plot medan vektor soal (b)
plt.figure(figsize=(8, 8))
plt.quiver(x, y, Fx_b, Fy_b, angles='xy', scale_units='xy', scale=1, color='red')
plt.title('Medan Vektor (b): $\\mathbf{F}(x, y) = y\\mathbf{i} + (x + y)\\mathbf{j}$', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$y$', fontsize=12)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```
